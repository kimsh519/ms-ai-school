```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Conv2D(32,(3,3),activation = 'relu',input_shape = (28,28,1))) # 흑백 이미지이기 때문에 1, color image = 3(RGB) 
model.add(layers.MaxPooling2D((2,2))) # Max pooling을 통해 (2,2)로 출력
model.add(layers.Conv2D(32,(3,3),activation = 'relu',input_shape = (28,28,1)))  
# 두 번째 레이어부터는 shape 가 줄어들어 input_shape를 따로 지정해 주지 않아도 적용된다.
# input_shape가 축소된 레이어의 shape 보다 크다면 빈 공간을 0으로 채워 주기 때문에 별 상관없지만, 더 작을 경우 문제가 생긴다
model.add(layers.Conv2D(32,(3,3),activation = 'relu',input_shape = (28,28,1)))  
model.add(layers.MaxPooling2D((2,2)))
```

```python
model.summary()
```

```python
model.add(layers.Flatten()) # 처리하기 좋은 형태의 레이어로 만든다
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
from keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

(train_images,train_labels),(test_images,test_labels) = mnist.load_data()
print(train_images.shape)
print(train_images[0])

# data scaling
train_data = train_images.reshape((60000,28,28,1)) #input shape와 차원을 맞춰주기 위해 차원을 늘림
print(train_data.shape)
print(train_data[0])
train_data = train_data.astype('float32')/255

test_data = test_images.reshape((10000,28,28,1))
test_data = test_data.astype('float32')/255

train_target = to_categorical(train_labels)  # 이진 벡터로 표현
test_target = to_categorical(test_labels)
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_data, train_target,epochs = 5, batch_size = 64)
```

```python
test_loss, test_acc = model.evaluate(test_data, test_target)
```
